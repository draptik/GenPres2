```polyglot-notebook
#load "load.fsx"

open Newtonsoft.Json

open FSharpPlus
open FSharpPlus.Data
open Informedica.Utils.Lib.BCL

open Informedica.OpenAI.Lib
```

```polyglot-notebook
let inline test extract print sysInput model texts =
    [
        for (text, exp) in texts do
            let act, msgs =
                text
                |> sysInput
                |> State.run
                    (extract model text)

            msgs |> print

            if act = exp then 1 
            else 
                printfn $"## Failed test: {act} is not {exp}"
                0
            , (act, exp)
            , text
    ]
```

```polyglot-notebook
let calcScore = List.sumBy (fun (s, _, _) -> s)


let testLocal =
    let sysInput text = 
        [
            Prompts.System.systemDoseQuantityExpert text
            |> Message.system
        ]

    [
//        Ollama.Models.llama2
//        Ollama.Models.gemma
//        Ollama.Models.openhermes
//        Ollama.Models.llama2
//        Ollama.Models.``llama-pro``
//        Ollama.Models.``openchat:7b``
//        Ollama.Models.``llama2:13b-chat``
        Ollama.Models.mistral
    ]
    |> List.map (fun model -> 
        printf $"- Testing: {model}: "
        let s = 
            Texts.testMinMaxDoseQtyTexts
            |> List.take 1
            |> test 
                Ollama.Extract.minMaxDose 
                (List.iter Message.print) 
                sysInput
                model
        printfn $"score: {s |> calcScore}"
        model, s
    )    
```

```polyglot-notebook
testLocal    
|> List.maxBy (fun (model, xs) -> xs |> calcScore)
|> fun (m, s) -> printfn $"\n\n## And the winner is: {m} with a high score: {s |> calcScore} from {Texts.testUnitTexts |> List.length}"

testLocal
|> List.collect (fun (m, xs) -> xs)
|> List.groupBy (fun (_, _, t) -> t)
|> List.map (fun (txt, s) -> s |> calcScore, txt)
|> List.sortBy fst
|> List.iter (fun (s, t) ->
    printfn $"## Score: {s}"
    printfn $"## Text:\n{t}"
)
```

```polyglot-notebook
let testChatGPT =

    [
        OpenAI.Models.``gpt-4-turbo-preview``
    ]
    |> List.map (fun model -> 
        let sysInput text =
            let msg =
                text
                |> Prompts.System.systemDoseQuantityExpert
                |> Message.system
            OpenAI.Chat.defaultChatInput model msg []
        printf $"- Testing: {model}: "
        let s = 
            Texts.testMinMaxDoseQtyTexts
            |> List.take 1 
            |> test 
                OpenAI.Extract.minMaxDose 
                OpenAI.Chat.print 
                sysInput
                model
        printfn $"score: {s |> calcScore}"
        model, s
    )    
```

```polyglot-notebook
testChatGPT    
|> List.maxBy (fun (model, xs) -> xs |> calcScore)
|> fun (m, s) -> printfn $"\n\n## And the winner is: {m} with a high score: {s |> calcScore} from {Texts.testUnitTexts |> List.length}"

testChatGPT
|> List.collect (fun (m, xs) -> xs)
|> List.groupBy (fun (_, _, t) -> t)
|> List.map (fun (txt, s) -> s |> calcScore, txt)
|> List.sortBy fst
|> List.iter (fun (s, t) ->
    printfn $"## Score: {s}"
    printfn $"## Text:\n{t}"
)
```
